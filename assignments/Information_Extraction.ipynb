```python
import spacy
from spacy import displacy

nlp = spacy.load("en_core_web_sm")

text = "I saw the man with a telescope."

doc = nlp(text)

for token in doc:
    print(
    f"{token.text:<15}"
    f"{token.dep_:<15}"
    f"{token.head.text}"
)
play = displacy.render(doc,style = "dep", jupyter = True)

display(play)
```

1. The Root is 'SAW'
2. The word 'telescope' modifies the man's position and how i am seeing the man.
3. Yes
4. Yes

```python
import spacy

nlp = spacy.load("en_core_web_sm")

text = """
Microsoft announced a new AI research initiative
in Berlin on 15 June 2026.

The project involves collaboration with Google,
SAP and SRH University Heidelberg.

Satya Nadella will present the initiative
at a conference in Germany.
"""

doc = nlp(text)

for ent in doc.ents:
    print(f"{ent.text:<20}{ent.label_:<20}{spacy.explain(ent.label_)}")
```

1. Entities like microsoft, Berlin, 15 june 2026, google, SAP, Satya nadella, germany were detected correctly and AI was incorrect.
2. labels like companies, agencies, institutions, countries, cities, states, dates, people.
3. the entity type 'ORG' has most often occured
4. yes, SRH, University and Heidelberg entities were missed out.

```python

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity

documents = [
    "AI is transforming healthcare",
    "Machine learning improves healthcare",
    "Football is a popular sport"
]

vectorizer = TfidfVectorizer()

tfidf_matrix = vectorizer.fit_transform(documents)

print("Vocabulary:")
print(vectorizer.get_feature_names_out())

print("\nTF-IDF Matrix:")
print(tfidf_matrix.toarray())

similarity_matrix = cosine_similarity(tfidf_matrix)

print("\nCosine Similarity Matrix:")
print(similarity_matrix)

print("Doc1 vs Doc2:", round(similarity_matrix[0][1], 3))
print("Doc1 vs Doc3:", round(similarity_matrix[0][2], 3))
print("Doc2 vs Doc3:", round(similarity_matrix[1][2], 3))
```

1. the highest similarity ois Doc1&Doc2 and Doc1&Doc3
2. Yes
3. Tf gives higher importance to the words that are unique to a document and appears less frequent.
4. No
5. No
